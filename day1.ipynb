```python
def chunk_text(text, chunksize=1500, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunksize
        chunks.append(text[start: end])
        start += chunksize - overlap
    return chunks
```

```python
import tiktoken
def chunk_text_by_tokens(text, max_tokens=500):
    # encoder = tiktoken.encoding_for_model(model)
    encoder = tiktoken.get_encoding('cl100k_base')
    tokens = encoder.encode(text)

    chunks = []

    for i in range(0, len(tokens), max_tokens):
        chunk_tokens = tokens[i:i + max_tokens]
        chunk_text = encoder.decode(chunk_tokens)
        chunks.append(chunk_text)
    
    return chunks
```

```python
def summarize_pdf(path):
    openai = OpenAI(
        base_url="http://localhost:11434/v1",
        api_key="ollama"  # required but can be any string
        )

    pdf_text = pdf_parser_func(path)
    # chunks = chunk_text(pdf_text)
    chunks = chunk_text_by_tokens(text=pdf_text)

    summaries = []
    for chunk in chunks:
        messages = [
            {'role':'system', 'content':'You are a helpful assitant who is academically excellent in summarizing texts'},
            {'role':'user', 'content':chunk}
        ]

        openai.chat.completions.create(
            model='tinyllama',
            messages=messages
        )
        summaries.append(response.choices[0].message.content)
    combined_summary = '\n'.join(summaries)

    final_messages = [
        {'role':'system', 'content':'You are a highly academic summarizer. Provide a clear, concise, and well-structured summary.'},
        {'role':'user', 'content': combined_summary}
    ]

    final_summary = openai.chat.completions.create(
        model='tinyllama',
        messages=final_messages
    )

    return final_summary
```
